### Hier die neu erstellte All Unterweight in die df Liste hinzufügen

```python
# 1. Underweight
df.loc[df["bmi"] < 18.5, 'Class_weights'] = 'Underweight'

# 2. Normal weight
df.loc[(df["bmi"] >= 18.5) & (df["bmi"] <= 24.9), 'Class_weights'] = 'Normal weight'

# 3. Overweight
df.loc[(df["bmi"] >= 24.91) & (df["bmi"] <= 29.9), 'Class_weights'] = 'Overweight'

# 4. xxl Overweight or Adipositas
df.loc[df["bmi"] >= 29.91 , 'Class_weights'] = 'xxl Overweight'
```
